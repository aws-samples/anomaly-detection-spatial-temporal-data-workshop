# find labels for the edge

```python
from tqdm import tqdm
```

```python
edge_label = {}
for idx, row in raw_net_trans_data.iterrows():
    if(idx%10000==0):
        print(idx)
    #print(row['Source'],row['Target'],row['fraud'])
    edge_label.setdefault((vertex_to_id[row['Source']],vertex_to_id[row['Target']]), []).append(row['fraud'])
```

```python
edge_label
```

```python
edge_label_unique = {k: np.unique(edge_label[k]) for k in edge_label}
```

```python
edge_label_unique
```

```python
edge_label_two_labels = {k: edge_label_unique[k] for k in edge_label_unique if len(edge_label_unique[k])>1 }
```

```python
edge_label_two_labels, len(edge_label_two_labels)
```

```python
edge_label_postprocessed = {k: max(edge_label_unique[k]) for k in edge_label_unique}
```

```python
len(edge_label), len(edge_label_unique),len(edge_label_two_labels), len(edge_label_postprocessed) #only 1108 out of 47132 edges had flipped labels 
```

```python
edge_label_postprocessed
```

```python
edge_label_postprocessed_df=pd.DataFrame.from_dict(edge_label_postprocessed,  orient='index', columns=['label'])
```
